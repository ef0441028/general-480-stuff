```python
from astropy.io import fits
import numpy as np
from matplotlib.colors import LogNorm
from photutils import CircularAnnulus as can
from photutils import aperture_photometry
from photutils import CircularAperture as ca
from astropy.table import QTable
from astropy.modeling import models
from astropy import units as u
from photutils.psf import (IterativelySubtractedPSFPhotometry,
                           BasicPSFPhotometry)
import pandas as pd
import matplotlib.pyplot as plt
```

```python
T = pd.read_csv('SQLstardata.csv')
```

```python
center_x = 132.82
center_y = 11.817

mask_inner = ((T['ra']-center_x)**2 
              + (T['dec'] - center_y)**2 
              < 0.05**2)

mask_outer = ((T['ra']-center_x)**2 
              + (T['dec'] - center_y)**2 
              > 0.05**2)

T_inner = T[mask_inner]
T_outer = T[mask_outer]
```

```python
u = T['u']
g = T['g']
r = T['r']
i = T['i']
z = T['z']

a = (u-g)
b = (r-i)
c = (r-z)
d = (g-i)
```

All Stars

```python
fig, ax = plt.subplots(2,2)                                               
fig.set_size_inches(11,8.5)                                               

fig.tight_layout()                                                       

ax[0,0].scatter(g,a , color='b', s=5)             
ax[0,1].scatter(r,b, facecolor='gray', s=5)                        
ax[1,0].scatter(a,b, facecolor='HotPink', s=5)      
ax[1,1].scatter(d,c, color='r', s=5); 

ax[1,1].set_xlim(0,3.5)
ax[0,0].invert_yaxis()
ax[0,1].invert_yaxis()
```
